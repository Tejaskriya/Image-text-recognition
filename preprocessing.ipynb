```python
import cv2
import numpy as np
import os

image_folder = r'D:\Final-yr-Proj\demoaimages' 
output_folder = r'D:\Final-yr-Proj\demoaimages\demooutput'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

image_paths = [os.path.join(image_folder, file) for file in os.listdir(image_folder)]
print('++',image_paths)
for image_path in image_paths:
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    print('==',image)
    ksize = 31
    sigma = 5
    theta = 0
    lamda = 10
    gamma = 0.5
    phi = 0
    kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, phi)
    if image is None:
        continue
    filtered_image = image
    kernel_dilate = np.ones((3, 3), dtype=np.uint8)
    dilated_image = cv2.dilate(filtered_image, kernel_dilate, iterations=1)

    text_region_mask = np.zeros_like(image, dtype=np.uint8)
    text_region_mask[dilated_image > 0] = 255
    text_region_filtered = cv2.bitwise_and(image, text_region_mask)

    output_file = os.path.join(output_folder, os.path.basename(image_path))
    cv2.imwrite(output_file, text_region_filtered)

    # Display original image, filtered image, dilated image, and text region filtered image
    #cv2.imshow("Original Image", image)
    #cv2.imshow("Filtered Image", filtered_image)
    #cv2.imshow("Dilated Image", dilated_image)
    #filtered_image = cv2.filter2D(image, cv2.CV_64F, kernel)
    #cv2.imshow("Text Region Filtered Image", text_region_filtered)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
